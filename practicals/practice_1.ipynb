```python
data_5 = {
    "Age": [25, 30, np.nan, 22, 28, np.nan],
    "Salary": [50000, 60000, 55000, np.nan, 52000, 58000],
    "Department": ["HR", "Finance", None, "IT", "Finance", "HR"],
    "Gender": ["Male", "Female", "Female", None, "Male", "Female"]
}

df = pd.DataFrame(data_5)

print(f"{separator} Data set: {separator}")
print(df)
print()
print(f"{separator} Missing Data: {separator}")
print(df.isnull().sum())

# df["Age"].fillna(df["Age"].median(), inplace=True)
df["Age"] = df["Age"].fillna(df["Age"].mean())
# df["Salary"].fillna(df["Salary"].median(), inplace=True)
df["Salary"] = df["Salary"].fillna(df["Salary"].median())
mode_value = df["Department"].mode()[0]
# df["Department"].fillna(mode_value, inplace=True)
df["Department"] = df["Department"].fillna(mode_value)
# df["Gender"].fillna(mode_value, inplace=True)
df["Gender"] = df["Gender"].fillna(mode_value)
print()
print(f"{separator} Missing Data after replacement: {separator}")
print(df.isnull().sum())
print()
print(f"{separator} Cleaned Data set: {separator}")
print(df)
```

```python
df = pd.DataFrame({"Age": ['25', '30', '22', '28']})

print(f"{separator} Before treating our data set: {separator}")
print(df)
print()
print(f">> Data type before conversion: \n=> {df['Age'].dtype}") # object (string)
print()
# Convert to numeric
df['Age'] = pd.to_numeric(df['Age'])
print(f">> Data type after conversion: \n=> {df['Age'].dtype}") # int64 (numeric)
print()
print(f"{separator} After treating our data set: {separator}")
print(df)
print()
print(df.mean())
print(divider)


df = pd.DataFrame({'Age': ['25', '30', 'unknown', '28']})

print(df.isnull().sum())
print()
# Convert to numeric, invalid 'unknown' becomes NaN
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
print(df)
print()
print(df.isnull().sum())
print()
df['Age'] = df['Age'].fillna(df['Age'].mean())
print(df.isnull().sum())
```

```python
data_6 = {
    "Age": ['25', '30', '22', '28', 'unknown'], # Numeric values as strings + invalid 'unknown'
    "Salary": ['50000', '60000', '55000', '52000', 'N/A'], # Numeric strings + 'N/A'
    "Department": ["HR", "Finance", "IT", "Finance", "HR"], # Categorical strings
    "StartDate": ['2020-01-15', '2019-07-10', '2021-03-22', '2020-11-05', 'unknown']  # Date strings + invalid
}

df = pd.DataFrame(data_6)
print(f"{separator} Original DataFrame: {separator}")
print(df)
print()
print(f"{separator} Missing Data: {separator}")
print(df.isnull().sum())
print()
print(f"{separator} Data Types: {separator}")
print(df.dtypes)
print(divider)
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
# df['Age'] = df['Age'].fillna(df['Age'].mean())

df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')
# df['Salary'] = df['Salary'].fillna(df['Salary'].mean())

df['Department'] = df['Department'].astype('category')
df['StartDate'] = pd.to_datetime(df['StartDate'], errors='coerce')

print(f"{separator} DataFrame after conversion: {separator}")
print(df)
print()
print(f"{separator} Data types after conversion: {separator}")
print(df.dtypes)
print()

df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
# df['StartDate'] = df['StartDate'].fillna(df['StartDate'].median())
df['StartDate'] = df['StartDate'].fillna(df['StartDate'].mode()[0])

print(f"{separator} DataFrame after treatment: {separator}")
print(df)
```

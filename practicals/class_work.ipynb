```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
separator = "_" * 20
divider = "\n" + "-"*100 + "\n"
```

```python
data = {
    "Student_ID": ['S1', 'S2', 'S3', 'S4', 'S5', 'S6', 'S7', 'S8', 'S9'],
    "Age": [20, 21, None, 19, 22, "twenty", 23, 24, None],
    "Score": [85, None, 78, 90, 88, 92, np.nan, 87, 80],
    "Hours_Student": [10, 15, 7, None, 12, 9, 14, 11, 8]
}

df = pd.DataFrame(data)
print(f"{separator} DataFrame: {separator}")
print(df)
```

```python
print(f"{separator} Shape: {separator}")
print(df.shape)
```

```python
print(f"{separator} Data types: {separator}")
print(df.dtypes)
```

```python
print(f"{separator} Missing Values: {separator}")
print(df.isnull().sum())
```

```python
print(f"{separator} Description: {separator}")
print(df.describe())
```

```python
print(f"{separator} Whole Information: {separator}")
print(df.info())
```

```python
print(f"{separator} Dataframe summary before Treatment {separator}")
print(df.info())
```

```python
# Treatment
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
print(f"{separator} Dataframe summary after Treatment {separator}")
print(df.info())
```

```python
print(f"{separator} DataFrame after treatment: {separator}")
print(df)
```

```python
print(f"{separator} Missing Values: {separator}")
print(df.isnull().sum())
```

```python
# Treating missing values
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Score"] = df["Score"].fillna(df["Score"].mean())
df["Hours_Student"] = df["Hours_Student"].fillna(df["Hours_Student"].mean())
print(f"{separator} Missing Values: {separator}")
print(df.isnull().sum())
```

```python
print(f"{separator} DataFrame after conversion: {separator}")
print(df)
```

```python
print(f"{separator} Graph: {separator}")

# Scatter plot
# plt.figure(figsize=(8, 5))
# plt.scatter(df['Hours_Student'], df['Score'], color='blue', edgecolor='black')

x = df['Hours_Student']
y = df['Score']
plt.scatter(x, y, color='green', marker='o')

# Title and labels
plt.title('Relationship Between Hours Studied and Exam Score')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')

# Grid for better readability
plt.grid(True)
plt.show()
```
